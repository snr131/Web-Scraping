```python
# import requests to get the html from the url
import requests
# import BeautifulSoup to parse the html
from bs4 import BeautifulSoup
```

```python
url = 'https://www.macrotrends.net/stocks/charts/AAPL/apple/stock-price-history'
```

```python
# get the url using the requests library
page = requests.get(url).text
```

```python
# convert the html to a BeautifulSoup object
soup = BeautifulSoup(page, features='lxml')
```

```python
table = soup.table
```

```python
table
```

```python
trs = table.find_all('tr')
```

```python
for tr in trs[2:]:
  byyear = tr.find_all('td')
  year = byyear[0].text
  yearClose = byyear[5].text
  print('''
  Year: {}
  Year Close: {}'''.format(year, yearClose))
```

```python
def get_aapl_byyear(str_url):
  page = requests.get(str_url).text
  soup = BeautifulSoup(page, features='lxml')
  table = soup.table
  print(table.find('th').text.strip())
  for tr in trs[2:]:
    byyear = tr.find_all('td')
    year = byyear[0].text
    yearClose = byyear[5].text
    print('''
    Year: {}
    Year Close: {}'''.format(year, yearClose))
```

```python
get_aapl_byyear('https://www.macrotrends.net/stocks/charts/AAPL/apple/stock-price-history')
```
